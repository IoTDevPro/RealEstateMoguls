### business license summary query

```python
select zipcode, extract(Mon from license_issue_date),count(*) 
from bbl_2017 group by zipcode, extract(Mon from license_issue_date) order by 1,2;

select zipcode,sum(license_issue_date)
from bbl_2017
where extract(Mon from license_issue_date) < 7 
group by zipcode 
order by 1 desc;

select zipcode, count(bbl_2017.license_issue_date) as licenses
from bbl_2017
where extract(Mon from license_issue_date) < 13
group by zipcode
order by 1 asc;

## this is for 6 month data

create sequence license_seq;
create table license_summary_float (license_sq integer, zipcode float, growth_YoY float, licenses_17 float, 
                              licenses_16 float,licenses_15 float ,licenses_14 float ,licenses_13 float, 
                              licenses_12 float);
ALTER TABLE license_summary ALTER COLUMN license_sq SET DEFAULT nextval('license_seq');

insert into license_summary_float (zipcode, licenses_17,licenses_16,licenses_15,licenses_14,licenses_13,licenses_12)
(
select z.zipcode, a17.licenses_17, a16.licenses_16, a15.licenses_15, a14.licenses_14, a13.licenses_13, a12.licenses_12
from 
(select distinct zipcode from zillow_weekly order by 1 asc) z,
(select bbl_2017.zipcode, count(*) as licenses_17
from bbl_2017
where extract(Mon from bbl_2017.license_issue_date) < 7
group by bbl_2017.zipcode
order by 1 asc) a17,
(select bbl_2016.zipcode, count(*) as licenses_16
from bbl_2016
where extract(Mon from bbl_2016.license_issue_date) < 7
group by bbl_2016.zipcode
order by 1 asc) a16,
(select bbl_2015.zipcode, count(*) as licenses_15
from bbl_2015
where extract(Mon from bbl_2015.license_issue_date) < 7
group by bbl_2015.zipcode
order by 1 asc) a15,
(select bbl_2014.zipcode, count(*) as licenses_14
from bbl_2014
where extract(Mon from bbl_2014.license_issue_date) < 7
group by bbl_2014.zipcode
order by 1 asc) a14,
(select bbl_2013.zipcode, count(*) as licenses_13
from bbl_2013
where extract(Mon from bbl_2013.license_issue_date) < 7
group by bbl_2013.zipcode
order by 1 asc) a13,
(select bbl_2012.zipcode, count(*) as licenses_12
from bbl_2012
where extract(Mon from bbl_2012.license_issue_date) < 7
group by bbl_2012.zipcode
order by 1 asc) a12
where z.zipcode = a17.zipcode
and z.zipcode = a16.zipcode
and z.zipcode = a15.zipcode
and z.zipcode = a14.zipcode
and z.zipcode = a13.zipcode
and z.zipcode = a12.zipcode);


ALTER TABLE license_summary_float add column growth_yoy_5 float;
ALTER TABLE license_summary_float add column growth_yoy_5 float;

insert into license_summary_float  (zipcode, growth_YoY)
(select zipcode, ((licenses_17 - licenses_16) / licenses_16)
                      from license_summary_float 
--where licensecategory = 'One Family Rental');

update license_summary_float
set growth_YoY = sq.yoy
from (select zipcode, ((licenses_17 - licenses_16) / licenses_16) as yoy
                      from license_summary_float ) as sq
where license_summary_float.zipcode = sq.zipcode;

update license_summary_float
set growth_yoy_5 = sq.yoy
from (select zipcode, ((licenses_17 - licenses_12) / licenses_12) as yoy
                      from license_summary_float ) as sq
where license_summary_float.zipcode = sq.zipcode;

select bbl_2017.zipcode, count(bbl_2017.license_issue_date) as license_17, 
count(bbl_2016.license_issue_date) as license_16,
count(bbl_2015.license_issue_date) as license_15,
count(bbl_2014.license_issue_date) as license_14,
count(bbl_2013.license_issue_date) as license_13,
count(bbl_2012.license_issue_date) as license_12
from bbl_2017,bbl_2016,bbl_2015,bbl_2014,bbl_2013,bbl_2012
where extract(Mon from bbl_2017.license_issue_date) < 7
and bbl_2017.zipcode = bbl_2016.zipcode
and bbl_2017.zipcode = bbl_2015.zipcode
and bbl_2017.zipcode = bbl_2014.zipcode
and bbl_2017.zipcode = bbl_2013.zipcode
and bbl_2017.zipcode = bbl_2012.zipcode
group by bbl_2017.zipcode
order by 1 asc;
```
